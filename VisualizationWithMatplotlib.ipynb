In this assignment, you will make some plots about Corona. First, download the Coronavirus Source Data.

First, filter the data for four countries (Spain, France, Germany, Italy) and the last 100 days.

Create a figure with two subplots. The left subplot will show total cases, and the right one will show total deaths for these countries

Set titles, X and Y labels, change the font, font size, and font color for each subplot.

Change color, width, and type of lines.

Put a legend to bottom right.

Change the color and rotation of the X-axis ticks.

And set a title for the figure (not for the plots).


```python
import pandas as pd

import matplotlib.pyplot as plt

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)


plt.figure(figsize=(20,10))
plt.suptitle('Corona Datas', fontsize=10)
spainDf = df[df['location'].isin(['Spain'])]
spain = spainDf.tail(100)
spainCases= pd.DataFrame(spain["total_cases"])
spainDeath = pd.DataFrame(spain["total_deaths"])

franceDf = df[df['location'].isin(['France'])]
france = franceDf.tail(100)
franceCases= pd.DataFrame(france["total_cases"])
franceDeath = pd.DataFrame(france["total_deaths"])


germanyDf = df[df['location'].isin(['Germany'])]
germany = germanyDf.tail(100)
germanyCases= pd.DataFrame(germany["total_cases"])
germanyDeath = pd.DataFrame(germany["total_deaths"])


italyDf = df[df['location'].isin(['Italy'])]
italy = italyDf.tail(100)
italyCases= pd.DataFrame(italy["total_cases"])
italyDeath = pd.DataFrame(italy["total_deaths"])




plt.subplot(1,2,1)
plt.plot(italy["date"],italyCases, c="red",label="Italy",linestyle='-',lw=1)
plt.plot(germany["date"],germanyCases, c="blue",label="Germany",linestyle='--',lw=2)
plt.plot(france["date"],franceCases, c="black",label="France",linestyle=':',lw=3)
plt.plot(spain["date"],spainCases, c="pink",label="Spain",linestyle='--',lw=4)
plt.xticks(rotation=90,  fontsize=9, color="#007D00")
plt.xlabel("Dates",fontsize=30,fontstyle='italic',color="#f15337")
plt.ylabel("Cases (million)",fontsize=30,fontstyle="italic",color="#f15337")
plt.title("Cases of Spain,Germany,Italy,France",fontsize=20,fontstyle="normal",color="#f15337")
plt.legend(loc=4,title="Countries")

plt.subplot(1,2,2)
plt.plot(italy["date"],italyDeath, c="red",label="Italy",linestyle='-',lw=1)
plt.plot(germany["date"],germanyDeath, c="blue",label="Germany",linestyle='--',lw=2)
plt.plot(france["date"],franceDeath, c="black",label="France",linestyle=':',lw=3)
plt.plot(spain["date"],spainDeath, c="pink",label="Spain",linestyle='--',lw=4)
plt.xticks(rotation=90,  fontsize=9 ,color="#007D00")
plt.xlabel("Dates",fontsize=30,fontstyle='oblique',color="#5337f1")
plt.ylabel("Deaths",fontsize=30,fontstyle='oblique',color="#5337f1")
plt.title("Deaths of Spain,Germany,Italy,France",fontsize=20,fontstyle="normal",color="#5337f1")
plt.legend(loc=4,title="Countries")
plt.tight_layout()


plt.show()
```
